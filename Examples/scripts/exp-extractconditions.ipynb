```python
# extract info such as temperature, pressure, equivalence ratio ranges, and paper dois
for exp in exps:
    id = str(exp.id)
    sps = '/'.join([sp.species.preferredKey for sp in exp.initial_species])
    Ts = '{:.1f}-{:.1f}'.format(exp.t_inf, exp.t_sup)
    Ps = '{:.1f}-{:.1f}'.format(exp.p_inf, exp.p_sup)
    PHIs = '{:.1f}-{:.1f}'.format(exp.phi_inf, exp.phi_sup)
    l = [str(prop.value)
         for prop in exp.common_properties if prop.name == 'length']
    if len(l) > 0:
        l = l[0]
    else:
        l = '-'
    t = [str(prop.value) for prop in exp.common_properties if prop.name == 'residence time']
    if len(t) > 0:
        t = t[0]
    else:
        t = '-'
    paper = exp.file_paper.reference_doi
    print('\t'.join([sps, 'JSR', Ts, Ps, PHIs, l, t, paper, id]))
```
