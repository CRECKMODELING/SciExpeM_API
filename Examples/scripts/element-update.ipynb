```python
specie = my_sciexpem.filterDatabase(
    model_name='Species', preferredKey = 'C10H6')
```

```python
specie = my_sciexpem.filterDatabase(
    model_name='Species', preferredKey='C10H6-14')
my_sciexpem.updateElement(
    element=specie[0], preferredKey='C6H4(C2H)2-14', names=['C10H6-14', 'C10H6', 'C6H4(C2H)2-14'])
```

```python
my_sciexpem.updateElement(element = specie, preferredKey='C6H13OH-1', CAS='111-27-3',
                          names = ['ESAN1OH', 'C6H13OH','C6H13OH-1'], InChI='1S/C6H14O/c1-2-3-4-5-6-7/h7H,2-6H2,1H3',
                          SMILES='CCCCCCO', chemName = '1-Hexanol')
```
